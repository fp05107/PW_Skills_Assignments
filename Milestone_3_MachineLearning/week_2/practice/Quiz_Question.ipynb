| **Goal**                                  | **Best Plot**                  | **Why**                                                   |
| ----------------------------------------- | ------------------------------ | --------------------------------------------------------- |
| **Distribution** (spread, skew, outliers) | **Histogram** / **Boxplot**    | Shows frequency and spread, detects skew/outliers         |
| **Compare groups** (e.g., by job, gender) | **Boxplot** or **Violin plot** | Shows medians, quartiles, and variation across categories |
| **Time-based trends** (monthly over time) | **Line chart**                 | Shows how income changes month to month                   |
| **Income ranges** (e.g., income brackets) | **Bar chart** or **Pie chart** | Visualizes proportions in each income group               |
| **Relationship with another variable**    | **Scatter plot**               | Helps find correlation (e.g., income vs. experience)      |


```python
plt.figure(figsize=(10, 6))
plt.hist(df['MonthlyIncome'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Monthly Income')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
