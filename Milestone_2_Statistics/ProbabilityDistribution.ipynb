```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Normal Distribution or Gaussian distribution
# Majority of data point is clustered around mean

mu = 0
sigma = 0.1

s = np.random.normal(mu, sigma, size = 10000)
```

```python
len(s)
```

```python
plt.hist(s, 30)
plt.show()
```

```python
# Uniform distribution
# All the outcomes has equal probability to occur in a range
u = np.random.uniform(0, 1, 10000)
```

```python
len(u)
```

```python
plt.hist(u)
plt.show()
```

```python
sns.displot(u)
```

```python
# Log normal distribution
l = np.random.lognormal(mean = 1, sigma = 0.5, size = 1000)
```

```python
l
```

```python
plt.hist(l, bins=100)
plt.show()
```

```python
sns.boxplot(l) # since lognormal is right skewed so outliers are present
```

```python
sns.distplot(l)
```

```python
#lognormal >> normal distribution if you take log of the values

sns.distplot(np.log(l))
```

```python
# square root transformation
sns.distplot(np.sqrt(l))
```

```python
# Chi square distribution
c = np.random.chisquare(df = 30, size = 10000)
c = c[np.where(abs(c) < 100)]
```

```python
plt.hist(c, bins=1000)
plt.show()
```
