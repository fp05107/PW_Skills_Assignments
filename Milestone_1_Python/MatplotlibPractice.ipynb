```python
x = [1, 2, 3, 4]
y = [1, 2, 3, 5]
plt.plot(x, y, color = 'red', marker = "o", linestyle = '--', linewidth = 3, markersize=8)
plt.title("Line chart to understand syntax")
plt.xlabel("Day Intrerval as x plot")
plt.ylabel("Y label")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 1, 3]

plt.plot(x, y1, color = 'green', marker = 'o', linewidth = 3, markersize = 8, label = 'line y1')
plt.plot(x, y2, color = 'b', marker = 'o', linewidth = 3, markersize = 8, label = 'line y2')
plt.grid()
plt.title("Multile line chart")
plt.xlabel("x label")
plt.ylabel("stock price")
plt.legend()
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))

plt.show()
```
